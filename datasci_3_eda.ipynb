## Looking for Missingness

### Cleaning White Spaces

```python
import re
def cleaning_columns (df):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

    df.columns = [clean_name(col) for col in df.columns]
    return df

df = cleaning_columns(df)
df
```

```python
df.columns
```

```python
df.dropna(subset=['2020hhstargets','county'],inplace=True)
df
```

## Handling Outliers

```python
catheterdays=df['catheterdays']
```

```python
q1 = np.percentile(catheterdays, 25)
q3 = np.percentile(catheterdays, 75)
iqr = q3 - q1
```

```python
threshold_lower = q1 - 1.5 * iqr
threshold_upper = q3 + 1.5 * iqr
```

```python
outliers = [x for x in catheterdays if x < threshold_lower or x > threshold_upper]
```

```python
print("Outliers:", outliers)
```

### Approaching Outliers

In this dataset I believe the outliers should be retained. In the context of this dataset, I would like to gain a better understaning between the amount of days a catheter is kept in a patient and its result in observed infections. A difference of a couple of days can have a huge impact on whether a patient is diagnosed with an infection or not, therefore it is crucial to retian the outliers as a major difference in observed infections may be noted at these values. For example hospitals with a higher catheter day count may be seen having a higher observed infection count, in which instance these outliers would be necessary to determine a relationship between these two variables.
